```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months =['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# Made-up monthly sales data
Wafers= np.array([5, 7, 8, 6, 7, 9, 10, 12, 11, 13, 14, 15])
Kurkure = np.array([3, 4, 5, 4, 6, 5, 7, 8, 7, 9, 10, 11])
Dark_Chocolate = np.array([2, 3, 4, 3, 4, 6, 5, 6, 5, 7, 8, 9])
plt.figure(figsize=(10, 6))
plt.stackplot(months, Wafers,Kurkure,Dark_Chocolate,
              labels=['Wafers','Kurkure','Dark_Chocolate'],
              colors=['yellow', 'orange', 'brown'])
plt.title('Cumulative Sales of Products Over 12 Months')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
