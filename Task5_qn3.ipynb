```python
import matplotlib.pyplot as plt
```

```python
categories=['Ganeshotsav','Diwali','Navratri','Christmas','Holi']
value=[99,89,75,57,60]
bars=plt.bar(categories,value,color='Blue')
plt.xlabel("<-----Festivals----->")
plt.ylabel("Popular Festivals of Mumbai")
plt.title("Popularity vise Graph of Festivals in (%)")
for i in range(len(categories)):
    plt.text(i, value[i]+1, str(value[i]), ha='center')
plt.ylabel('Percentage')
plt.title('Bar Chart with Annotations')
plt.show()
plt.tight_layout()
```
