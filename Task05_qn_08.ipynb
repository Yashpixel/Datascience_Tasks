```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sub = {
    'Python': np.random.randint(50, 100, 20),
    'C': np.random.randint(40, 95, 20),
    'Java': np.random.randint(45, 90, 20)
}
df = pd.DataFrame(sub)
plt.figure(figsize=(8, 5))
df.boxplot(column=['Python', 'C', 'Java'], grid=False, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='blue'),
           medianprops=dict(color='red'))
plt.title('Student Exam Scores Distribution')
plt.ylabel('Scores')
plt.tight_layout()
plt.show()

```
