```python
import matplotlib.pyplot as plt
```

```python
companies = ['Mahindra', 'Hyundai', 'Toyota', 'Tesla']
market_share = [35, 25, 20, 20]
largest_index = market_share.index(max(market_share))
explode = [0.1 if i == largest_index else 0 for i in range(len(market_share))]
plt.figure(figsize=(7, 7))
plt.pie(market_share, labels=companies, autopct='%1.1f%%',
        explode=explode, startangle=90, colors=['skyblue', 'lightgreen', 'salmon', 'orange'],
        shadow=True)
plt.title('Market Share of Companies')
plt.tight_layout()
plt.show()
```
