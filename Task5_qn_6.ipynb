```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 50, 10)
y = np.sin(x)
categories = ['C', 'C++', 'Java', 'DBMS']
values = [1, 13, 2, 47]
scatter_x = np.random.rand(100)
scatter_y = np.random.rand(100)
hist_data = np.random.randn(1000)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(x, y, color='red', linestyle='-', marker='o')
axs[0, 0].set_title('Line Plot')

axs[0, 1].bar(categories, values, color='green')
axs[0, 1].set_title('Bar Chart of no. of failed students in subjects')

axs[1, 0].scatter(scatter_x, scatter_y, color='blue', alpha=0.7)
axs[1, 0].set_title('Scatter Plot')

axs[1, 1].hist(hist_data, bins=10, color='purple', edgecolor='black')
axs[1, 1].set_title('Histogram')

fig.suptitle('2x2 Subplots Grid Example', fontsize=16)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
