```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,400)
ysin=np.sin(x)
ycos=np.cos(x)
yexp=np.exp(x/5)
plt.figure(figsize=(10, 5))
plt.plot(x, ysin, label='sin(x)', color='red', linestyle='-')
plt.plot(x, ycos, label='cos(x)', color='green', linestyle='--')
plt.plot(x, yexp, label='exp(x / 5)', color='blue', linestyle=':')

plt.title('Multiple Lines in One Plot: sin, cos, exp')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
